### 1. Memilih Data untuk Pemodelan

Kita akan mulai dengan memilih beberapa variabel. Selanjutnya akan menunjukkan teknik statistik untuk memprioritaskan variabel secara otomatis.

Untuk memilih variabel/kolom, kita perlu melihat daftar semua kolom dalam kumpulan data. Itu dilakukan dengan properti kolom dari DataFrame (baris bawah kode di bawah).

```python
import pandas as pd

melbourne_file_path = 'https://raw.githubusercontent.com/AnggaPradiktas/data-raw-dump/main/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path) 
melbourne_data.columns
```

```python
melbourne_data.head()
```

```python
# Data Melbourne memiliki beberapa nilai yang hilang (beberapa rumah yang beberapa variabelnya tidak dicatat.)
# Kita akan belajar menangani nilai yang hilang di tutorial selanjutnya.
# Jadi kami akan mengambil opsi paling sederhana untuk saat ini, dan menghapus rumah dari data.

# dropna menjatuhkan nilai yang hilang (anggap na sebagai "tidak tersedia")
melbourne_data = melbourne_data.dropna(axis=0)
```

### 2. Memilih Target Prediksi
Kita dapat mengeluarkan variabel tertentu sebagai target. Kolom ini disimpan dalam bentuk Series, yang secara umum seperti DataFrame dengan hanya satu kolom data.

```python
y = melbourne_data.Price
```

```python
print(y)
```

### 3. Choosing "Features"
Kolom yang dimasukkan ke dalam model kita (dan kemudian digunakan untuk membuat prediksi) disebut "Features". Dalam kasus kami, itu akan menjadi kolom yang digunakan untuk menentukan harga rumah. Terkadang, Anda akan menggunakan semua kolom kecuali target sebagai fitur. Di lain waktu Anda akan lebih baik dengan lebih sedikit fitur.

Untuk saat ini, kita akan membuat model dengan hanya beberapa fitur. Nanti Anda akan melihat cara mengulangi dan membandingkan model yang dibuat dengan fitur berbeda.



```python
# Telah ditambahkan features sesuai soal: 'Car', 'Yearbuilt', 'BuildingArea'
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude', 'Car', 'YearBuilt', 'BuildingArea']
```

```python
X = melbourne_data[melbourne_features]
```

```python
X.describe()
```

```python
X.head()
```


### 4. Building Your Model
Anda akan menggunakan scikit-learn untuk membuat model Anda. Saat coding, library ini ditulis sebagai sklearn, seperti yang akan Anda lihat dalam contoh kode. Scikit-learn dengan mudah merupakan perpustakaan paling populer untuk memodelkan jenis data yang biasanya disimpan di DataFrames.

Langkah-langkah untuk membangun dan menggunakan model adalah:

* Define: Apa jenis modelnya? Decision Tree? Beberapa jenis model 
lainnya? Beberapa parameter lain dari tipe model juga ditentukan.
* Fit: Menangkap pola dari data yang disediakan. Ini adalah jantung dari pemodelan.
* Prediksi: Seperti apa penampakannya
* Evaluasi: Tentukan seberapa akurat prediksi model.

Berikut adalah contoh pendefinisian model pohon keputusan dengan scikit-learn dan menyesuaikannya dengan fitur dan variabel target

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, y)
```

Banyak model machine learning memungkinkan beberapa randomness dalam pelatihan model. Menentukan nomor untuk random_state memastikan Anda mendapatkan hasil yang sama di setiap proses. Ini dianggap sebagai praktik yang baik. Anda menggunakan nomor apa pun, dan kualitas model tidak akan sepenuhnya bergantung pada nilai yang Anda pilih.

Kita sekarang memiliki model yang dapat kita gunakan untuk membuat prediksi.

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```
